```python
X = [
    [1, 2, 3],
    [4, 5, 6],
    [6, 7, 8],
    [1, 0, 4],
]
y = [6, 15, 21, 5]
```

```python
from nn import MLP

mlp = MLP(3, [4, 1])

epochs = 1000
learning_rate = 0.001
for epoch in range(epochs):
    # forward pass
    y_pred = [mlp(x) for x in X]
    loss = sum([(yt - yp[0]) ** 2 for yt, yp in zip(y, y_pred)])

    # zero-grad
    for p in mlp.params():
        p.grad = 0

    # backward pass
    loss.backward()
    for p in mlp.params():
        p.data = p.data - learning_rate * p.grad

    print(f'Epoch {epoch} loss: {loss.data}')
```
